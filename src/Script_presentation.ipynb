# Training the model

```python
import datareader as dr
import word2vec as wv
import time
import util
```

```python
ts = time.time()
file_path = "./data/pt96.txt"
my_data = dr.DataReader(file_path)
my_data.get_data()
word2index = my_data.word2index
index2word = my_data.index2word
config = wv.Config()
my_model = wv.SkipGramModel(config)
embeddings = wv.run_training(my_model, my_data)
te = time.time()
```

```python
#! tensorboard ..
```

```python
#! tensorboard  --logdir=./processed
```

```python
print("Duration of the training = {}".format(te-ts))
```

# Evaluating the model

```python
eval_path = "./evaluation/questions-words-ptbr.txt"
score, report = util.score(index2word, word2index, embeddings, eval_path,verbose=False,raw=True)
```

```python
print("Overall score ={}".format(score))
print("Detailed score:")
for result in report:
    print(result)
```

```python
import gensim, logging
import os
import inspect
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)


class MySentences(object):
    def __init__(self, dirname):
        self.dirname = dirname
 
    def __iter__(self):
        for fname in os.listdir(self.dirname):
            for line in open(os.path.join(self.dirname, fname)):
                yield line.split()

file_path = "./data/gensim"
sentences = MySentences(file_path)
model = gensim.models.Word2Vec(sentences)
```
